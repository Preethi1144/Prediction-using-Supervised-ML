```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
url = "http://bit.ly/w-data"
dataset = pd.read_csv(url)
print("Data imported")
```

```python
sns.regplot(data=dataset, x='Hours',y='Scores', ci=None )
plt.title('Dataset plot')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
#Preparing data
x = dataset['Hours'].values.reshape(-1,1)
y = dataset['Scores'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Training model
reg = LinearRegression()
reg.fit(x_train, y_train)
print("Model trained")
```

```python
#Predict values
y_val = reg.predict(x_test)
y_test.shape
```

```python
data_df = pd.DataFrame({'Actual values': y_test.flatten(), 'Predicted values': y_val.flatten()})
data_df
```

```python
hours = np.array(9.25).reshape(-1,1)
predicted = reg.predict(hours)
print(f'''Number of hours :{hours.flatten()[0]}
     Marks predicted: {predicted.flatten()[0]}''')
```

```python
#EValuate model
rootmeansq = np.sqrt(mean_squared_error(y_test,y_val))
print("Root mean squared error = {}".format(rootmeansq))
```
